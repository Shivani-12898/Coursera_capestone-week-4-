Introduction:

The main purpose of this project is to basically help and find out the best location suitable to an indiviual that is planning to move out of town becuase of job. 

Now, when any individual plans to move out there are a number factors that are considered by him/her for example schools, super markets,cafe, grocery shops etc so that its easy to adjust at the new place. 

Therefore, the basic aim of this project is to now and comapre different neighbour hoods before moving out and deciding where to buy a house. To create an analysis of features for a people migrating to Scarborough, Canada. 

It will help people to choose amongst the various ones available that is best for them according to there preferences.
